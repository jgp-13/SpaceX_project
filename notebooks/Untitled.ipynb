```python
import requests
import pandas as pd
import numpy as np
import datetime
import os
import space_x


from space_x.config import RAW_DATA_DIR, INTERIM_DATA_DIR, PROCESSED_DATA_DIR

spacex_file = os.path.join(INTERIM_DATA_DIR,'Spacex.csv')
part2_file = os.path.join(INTERIM_DATA_DIR, 'dataset_part_2.csv')

pd.set_option('display.max_columns', None)    # Setting this option will print all collumns of a dataframe
pd.set_option('display.max_colwidth', None)   # Setting this option will print all of the data in a featureimport pandas as pd

# Load both datasets
df_original = pd.read_csv(spacex_file)
df_part2 = pd.read_csv(part2_file)

# 1. Compare Column Names
print("Columns in Spacex.csv:")
print(df_original.columns.tolist())
print("\nColumns in dataset_part_2.csv:")
print(df_part2.columns.tolist())

# Identify differences in column names
original_cols = set(df_original.columns)
part2_cols = set(df_part2.columns)
print("\nColumns in Spacex.csv not in dataset_part_2.csv:", original_cols - part2_cols)
print("Columns in dataset_part_2.csv not in Spacex.csv:", part2_cols - original_cols)

# 2. Compare Data Types
print("\nData types in Spacex.csv:")
print(df_original.dtypes)
print("\nData types in dataset_part_2.csv:")
print(df_part2.dtypes)

# 3. Compare Shape (number of rows and columns)
print("\nShape of Spacex.csv:", df_original.shape)
print("Shape of dataset_part_2.csv:", df_part2.shape)

# 4. Compare Summary Statistics (for numerical columns)
print("\nSummary statistics for Spacex.csv:")
print(df_original.describe(include='all'))
print("\nSummary statistics for dataset_part_2.csv:")
print(df_part2.describe(include='all'))

# 5. Preview the first few rows
print("\nFirst few rows of Spacex.csv:")
print(df_original.head())
print("\nFirst few rows of dataset_part_2.csv:")
print(df_part2.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Define the available Seaborn styles
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

# Create a figure with one subplot per style
fig, axs = plt.subplots(1, len(styles), figsize=(20, 4))

# Loop through each style and corresponding subplot
for ax, style in zip(axs, styles):
    # Set the current style
    sns.set_style(style)
    # Plot the sample data
    ax.plot(x, y, color='b')
    # Set the title as the style name
    ax.set_title(style, fontsize=12)
    # Label the axes
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')

# Add a main title to the figure
plt.suptitle("Comparison of Seaborn Styles", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
